Simulate paths for the underlying share and for the counterparty’s firm value using sample sizes of 1000, 2000, …, 50000.
Using monthly simulations for the lifetime of the option.

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import stats, random
from scipy.stats import norm
 
S_0 = 100
risk_free_rate = 0.08
sigma = 0.3
K  = 120
barrier = 150
n_steps = 12
n_simulations = 50
T = 1
dt = T/n_steps

V_0 = 210
firm_sigma = 0.25
recovery_rate = 0.25
debt = 175


def asset_price(S,risk_free_rate,sigma,T,Z):
   return S*np.exp((risk_free_rate-(sigma**2)/2)*T + sigma*math.sqrt(T)*Z)

def call_payoff(S_T, K):
    return np.maximum(S_T-K,0)

def price_path(n_simulatios,n_steps,risk_free_rate,sigma,T,S_0,x,y,text):
    np.random.seed(0)
    path_matrix = np.zeros((n_steps,1))
    path_matrix[0,0] = S_0 #Initial value 
    
    fig = plt.figure()
    ax1 = fig.add_subplot(211, ylabel="Share price in $",xlabel="Sample Size")
    plt.plot(ax=ax1)
    
    for index in range(1,n_simulations+1):
        random_variable = np.random.randn(1000*index)
        for i in range(n_steps-1):
            path_matrix[i+1,0] = asset_price(path_matrix[i,0],risk_free_rate,sigma,T,random_variable[i])
        plt.plot(path_matrix[:,0],label="dfjlkj")
    plt.text(x,y,text)
    plt.show()
    
price_path(n_simulations,n_steps,risk_free_rate,sigma,dt,S_0,0,190,"Price path for Underying Shares") 
price_path(n_simulations,n_steps,risk_free_rate,firm_sigma,dt,V_0,0,320,"Price path for Counterparty’s firm value")
```

Determine Monte Carlo estimates of both the default-free value of the option and the Credit Valuation Adjustment (CVA).

```python
def black_scholes_call(S,X,T,rf,sigma):
    """
       Returns: Call value under Black-Schole-Merton option model
       Format   : bs_call(S,X,T,r,sigma)
               S: current stock price
               X: exercise price
               T: maturity date in years
              rf: risk-free rate (continusouly compounded)
           sigma: volatiity of underlying security 
    """    
    d1 = (np.log(S/X)+(rf+sigma**2/2)*T)/(sigma*np.sqrt(T))
    d2 = d1-sigma*np.sqrt(T)
    return S*norm.cdf(d1)-X*np.exp(-rf*T)*stats.norm.cdf(d2)
```

```python
def up_and_out_call_option(s0,x,T,r,sigma,n_simulation,barrier):
    """
        Returns: Call value of an up-and-out barrier option with European call
    """
    n_steps=12 # Define number of steps.
    dt = T/n_steps
    total=0
    for j in range(0,n_simulation):
        sT=s0
        out=False
        for i in range(0,int(n_steps)):
            e= random.normal()
            sT*= np.exp((r-0.5*sigma**2)*dt+sigma*e*np.sqrt(dt))
            if sT>barrier:
                out=True
        if out==False:
            total+=black_scholes_call(s0,x,T,r,sigma)
    return total/n_simulation
```

```python
def default_probability(v_0,firm_debt,T,rf,sigma_firm):
    d1 = (np.log(v_0/firm_debt)+(rf+sigma_firm**2/2)*T)/(sigma_firm*np.sqrt(T))
    d2 = d1-sigma_firm*np.sqrt(T)
    return norm.cdf(-d2)
```

```python
n_simulations = 50000 
result = up_and_out_call_option(S_0,K,T,risk_free_rate,sigma,n_simulations,barrier)
print('Price for the Up-and-out Call = ', round(result,3))
```

Calculate the Monte Carlo estimates for the price of the option incorporating counterparty risk, given by the default-free price less the CVA.

```python
np.random.seed(0)
simulations = 50000
corr_tested = np.linspace(-1,1,21)
cva_estimates = [None]*len(corr_tested)
cva_std = [None]*len(corr_tested)

for i in range(len(corr_tested)):
    correlation = corr_tested[i]
    if (correlation == 1 or correlation == -1):
        norm_vec_0 = norm.rvs(size=simulations)
        norm_vec_1 = correlation*norm_vec_0
        corr_norm_matrix = np.array([norm_vec_0, norm_vec_1])
    else:
        corr_matrix = np.array([[1, correlation],[correlation,1]])
        norm_matrix = norm.rvs(size = np.array([2, simulations]))
        corr_norm_matrix = np.matmul(np.linalg.cholesky(corr_matrix), norm_matrix)
        
    term_stock_val = asset_price(S_0,risk_free_rate,sigma,T,corr_norm_matrix[0,]) 
    call_val = call_payoff(term_stock_val,K)
    term_firm_val = asset_price(V_0,risk_free_rate,firm_sigma,T,corr_norm_matrix[1,]) 
    amount_lost = np.exp(-risk_free_rate*T)*(1-recovery_rate)*(term_firm_val < debt)*call_val
    cva_estimates[i] = np.mean(amount_lost)
    cva_std[i] = np.std(amount_lost)/np.sqrt(simulations)
```

```python
analytical_call_price = black_scholes_call(S_0,K,T,risk_free_rate,sigma)
default_prob = default_probability(V_0,debt,T,risk_free_rate,firm_sigma)
uncorr_cva = (1-recovery_rate)*default_prob*analytical_call_price
plt.plot(corr_tested, [uncorr_cva]*21)
plt.plot(corr_tested, cva_estimates, '.')
plt.plot(corr_tested, cva_estimates+3*np.array(cva_std), 'r')
plt.plot(corr_tested, cva_estimates-3*np.array(cva_std), 'b')
plt.xlabel('Correlation')
plt.ylabel('CVA')
plt.show()
```
